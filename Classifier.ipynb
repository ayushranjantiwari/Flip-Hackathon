# Balancing the Unbalanced Data

```python
df["Classification"].value_counts()
```

**Here the data is unbalanced, so we are going to use SMOTE**

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from imblearn.under_sampling import RandomUnderSampler
```

```python
df['sdsd'][df["sdsd"]=='--'] = 'NAN'
```

```python
df['sdsd'] = df['sdsd'].astype('float')
```

```python
df['sdsd'] = df['sdsd'].fillna(df['sdsd'].mean())
```

**Splitting the data into dependent and independent features**

```python
x = df.iloc[:,:21]
y = df["Classification"]
```

```python
y.value_counts()
```

```python
y_modify = []
for i in y:
    if (i=='A'):
        y_modify.append(1)
    else:
        y_modify.append(0)
        
y_modify = np.array(y_modify)
```

# Train Test Split 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y_modify,test_size=0.33,random_state=120)
```

```python
over = SMOTE(sampling_strategy=0.5)
under = RandomUnderSampler(sampling_strategy=0.5)
steps = [("o", over), ("u", under)]
pipeline = Pipeline(steps=steps)
x_res_train, y_res_train = over.fit_resample(x_train, y_train)
```

# Feature Engineering

**If we are using distance based algorithms, scaling is must. so we are going to do feature scaling now**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_res_train)
x_test_scaled = scaler.transform(x_test)
```

# Model Building

# ANN Classifier

```python
import tensorflow as tf
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
import tensorflow as tf
tf.random.set_seed(42)


model = tf.keras.Sequential([
    tf.keras.layers.Dense(21, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(3, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    loss=tf.keras.losses.binary_crossentropy,
    optimizer=tf.keras.optimizers.Adam(lr=0.03),
    metrics=[
        tf.keras.metrics.BinaryAccuracy(name='accuracy'),
        tf.keras.metrics.Precision(name='precision'),
        tf.keras.metrics.Recall(name='recall')
    ]
)

history = model.fit(x_train_scaled, y_res_train, epochs=200)
```

```python
# Test Model
from sklearn.metrics import classification_report, confusion_matrix

predY = model.predict(x_test_scaled)
y_pred= predY>.6
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
from sklearn.metrics import f1_score
print("F1-SCORE is: ",f1_score(y_test, y_pred))
```
